```python
import torch
import torch.nn as nn
```

```python
X_norm_train_np = X_norm_train.to_numpy()
Y_train_np = Y_train.to_numpy()
X_norm_train_tensor = torch.tensor(X_norm_train_np).float()
Y_train_tensor = torch.tensor(Y_train_np).float()

X_norm_test_np = X_norm_test.to_numpy()
Y_test_np = Y_test.to_numpy()
X_norm_test_tensor = torch.tensor(X_norm_test_np).float()
Y_test_tensor = torch.tensor(Y_test_np).float()
```

```python
X_train_np = X_train.to_numpy()
Y_train_np = Y_train.to_numpy()
X_train_tensor = torch.tensor(X_train_np).float()
Y_train_tensor = torch.tensor(Y_train_np).float()

X_test_np = X_test.to_numpy()
Y_test_np = Y_test.to_numpy()
X_test_tensor = torch.tensor(X_test_np).float()
Y_test_tensor = torch.tensor(Y_test_np).float()
```

```python
print(X_train_tensor.shape)
print(Y_train_tensor.shape)
print(X_test_tensor.shape)
print(Y_test_tensor.shape)
```

```python
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(10,1)
        self.fc2 = nn.Linear(1, 1)
        self.sigmoid = torch.nn.Sigmoid()

    def forward(self, x):
        hidden = self.fc1(x)
        hidden = self.sigmoid(hidden)
        output = self.fc2(hidden)
        output = self.sigmoid(output)
        return output
```

```python
model = MLP()
```

```python
import torch.optim as optim

criterion = nn.BCELoss() #reduction = mean
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
model.eval()
Y_pred = model(X_norm_test_tensor)
print(Y_pred.shape)
before_train = criterion(Y_pred, Y_test_tensor)
print('Test loss before training' , before_train.item())
```

```python
model.eval()
Y_pred = model(X_test_tensor)
print(Y_pred.shape)
before_train = criterion(Y_pred, Y_test_tensor)
print('Test loss before training' , before_train.item())
```

```python
model.train()
epoch = 10
for epoch in range(epoch):
    optimizer.zero_grad()
    Y_pred = model(X_norm_train_tensor)
    loss = criterion(Y_pred, Y_train_tensor)
    print('Epoch {}: train loss: {}'.format(epoch, loss.item()))
    loss.backward()
    optimizer.step()
```

```python
model.train()
epoch = 10
for epoch in range(epoch):
    optimizer.zero_grad()
    Y_pred = model(X_train_tensor)
    loss = criterion(Y_pred, Y_train_tensor)
    print('Epoch {}: train loss: {}'.format(epoch, loss.item()))
    loss.backward()
    optimizer.step()
```

```python
model.eval()
Y_pred = model(X_norm_test_tensor)
after_train = criterion(Y_pred, Y_test_tensor)
print('Test loss after training' , after_train.item())
```
